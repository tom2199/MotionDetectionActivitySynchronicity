# 2. Data Import


As an example here is the loading of the data from thursday (swimming) from the bangle with the mac adress 06d6.

```python
from scripts.preprocessing import *

# Decompress binary data
print("Decompressing data...")
data = readBinFile("data/bin/Thursday/06d6_d202209291636.bin")
dataframe = decompress(data)
print("Decompressing done.")

# Make the dataframe equidistant
print("Dataframe: make_equidistant...")
dataframe_equidistant = make_equidistant([dataframe], 25)[0][0]
print("make_equidistant done.")

# Resample dataframe
print("Dataframe: resample_raw_data...")
dataframe_resampled = resample_raw_data(dataframe_equidistant, 25)
print("resample_raw_data done.")

# Print the first rows of the dataframe
dataframe_resampled.head(10)
```

# 3. Visualize Data

Now we visualize the data to inspect it.

```python
import plotly.graph_objects as go
import numpy as np

# assign variables to the activity name, sensor names, and sampling rate (50)
sensor_names = ['x_axis', 'y_axis', 'z_axis']
sampling_rate = 25

def plot_activity(data, sensor_names, sampling_rate):
    
    # define the y- and x-axis as defined above
    # count for x how many records there are, divide by sampling rate
    y_axis = data[sensor_names]
    x_axis = np.array(range(len(data))) / sampling_rate

    # plot data
    fig = go.Figure()
    for s in sensor_names: 
      fig.add_trace(go.Scatter(x=x_axis, y=y_axis[s], name=s))
    fig.update_layout(margin_l=0,margin_r=0)  # no margins left or right
    fig.update_layout(xaxis_title="time (seconds)")
    fig.update_layout(yaxis_title="acceleration (10g)")
    fig.update_layout(legend_x=0, legend_bgcolor='rgba(0,0,0,0.2)')
    fig.show()

# call the function you just defined using the correct inputs
plot_activity(dataframe_resampled, sensor_names, sampling_rate)
```
