# 2. Visualization of data

For the batch processing of data I've written a few functions which we can use here to display the data of all stair climbing activity. Because of the large dataset I only display the necessary dataframes. Even with this, the plot would not load so I added plotly_resampler which resamples the data dynamically while displaying the plot(s).

```python
import pickle
from plotly.subplots import make_subplots
from plotly_resampler import FigureResampler
from scripts import processing


# open the dictionary of friday data (saved as a pickle file)
with open("data/pkl/friday.pkl", "rb") as f:
  jogging = pickle.load(f)
  f.close()

sensor_names = ["x_axis", "y_axis", "z_axis"]
# use FigureResampler from plotly_resampler to display all figures the dict contains
fig = FigureResampler(make_subplots(rows=len(jogging), cols=1, shared_xaxes=True, shared_yaxes="all", vertical_spacing = 0.015, subplot_titles=list(jogging.keys())))

for index, item in enumerate(jogging):
    dataframe_list = processing.return_dataframe_list_within_timerange(jogging.get(item), "2022-09-30 05:30", "2022-09-30 06:40")
    # show legend only for first graph (so there's only one legend)
    if index == 0:
      traces = processing.trace_dataframe_list(dataframe_list, sensor_names, True)
    else:
      traces = processing.trace_dataframe_list(dataframe_list, sensor_names, False)
    for trace in traces:
      fig.add_trace(trace, row=index+1, col=1)

fig.update_layout(margin_l=0, margin_r=0, height=1600)  # no margins left or right, bigger height
fig.update_layout(yaxis_title="acceleration")

# show figure
fig.show_dash()
```

# 3. Analysis of Attributes

Calculation of the chosen attributes within 10s windows

```python
import pandas as pd

jogging_vec = dict()

# concat dataframe lists from all bangles and select the timeframe for all of them
for key in jogging:
  df = pd.concat(jogging[key])
  jogging_vec[key] = df["2022-09-30 05:30":"2022-09-30 06:40"]

# get vector_magnitude for all items
for key in jogging_vec:
  jogging_vec[key] = processing.calc_vec_mag(jogging_vec[key])

# calculate rolling attributes for all items
jogging_rolling_10s = dict()
for key in jogging_vec:
  df = jogging_vec[key]
  jogging_rolling_10s[key] = processing.calc_rolling_attributes(df, ["min", "max", "mean", "std", "diff", "fft_freq", "quant0.2", "quant0.5", "quant0.8", "iqr0.2", "iqr0.5", "iqr0.8"], timeframe="10s")
```

Short script which compares the mean of each attribute within the activity range and outside the activity range

```python
import numpy as np

dictionary = jogging_rolling_10s
result_pre = pd.DataFrame()
result_post = pd.DataFrame()
result = pd.DataFrame()
rows = ["min", "max", "mean", "std", "diff", "fft_freq", "quant0.2", "quant0.5", "quant0.8", "iqr0.2", "iqr0.5", "iqr0.8"]
result_pre.loc[:, rows] = np.nan
result_post.loc[:, rows] = np.nan
result.loc[:, rows] = np.nan

start = "2022-09-30 05:40:00"
end = "2022-09-30 06:33:00"

for key in dictionary:
  df = dictionary[key]
  df_activity = df.loc[start:end]
  df_pre_activity = df.loc[:start]
  df_post_activity = df.loc[end:]
  result_pre.loc[''.join(key)] = df_pre_activity.mean()
  result_post.loc[''.join(key)] = df_post_activity.mean()
  result.loc[''.join(key)] = df_activity.mean()

result_combined = (result_pre.mean() + result_post.mean())/2
print((result.mean().abs() - result_combined.abs()).abs())
```

Plot of all dataframe attributes

```python
dictionary = jogging_rolling_10s
sensor_names = ["min", "max", "mean", "std", "diff", "fft_freq", "quant0.2", "quant0.5", "quant0.8", "iqr0.2", "iqr0.5", "iqr0.8"]

# use FigureResampler from plotly_resampler to display all figures the dict contains
fig = FigureResampler(make_subplots(rows=len(dictionary), cols=1, shared_xaxes=True, shared_yaxes="all", vertical_spacing = 0.015, subplot_titles=list(dictionary.keys())))

for index, item in enumerate(dictionary):
    dataframe = dictionary.get(item)
    # show legend only for first graph (so there's only one legend)
    if index == 0:
      traces = processing.trace_dataframe_list([dataframe], sensor_names, True)
    else:
      traces = processing.trace_dataframe_list([dataframe], sensor_names, False)
    for trace in traces:
      fig.add_trace(trace, row=index+1, col=1)

fig.update_layout(margin_l=0, margin_r=0, height=1600)  # no margins left or right, bigger height
fig.update_layout(yaxis_title="acceleration")

# show figure
fig.show_dash()
```

# 4. Activity Detection

Calculate score from iqr (with removed outliers), use activity detection as mask for score and then calculate identifier

```python
dictionary = jogging_rolling_10s

for key in dictionary:
  df = dictionary[key]
  df["iqr"] = processing.remove_outliers_by_zscore(df["iqr0.8"], 3)
  df["activity"] = processing.calc_activity(df["diff"])
  df["score"] = processing.calc_score(df["iqr"], 1, 1.2, resample="60s", timeframe="10s")
  df["score_clean"] = df["activity"] * df["score"]
  df["identifier"] = processing.calc_identifier(df["score_clean"], gap=60)
  dictionary[key] = df
```

Plot data (score, activity mask, score_clean (score * activity), identifier)

```python
sensor_names = ["score", "activity", "score_clean", "identifier"]
fig = make_subplots(rows=len(dictionary), cols=1, shared_xaxes=True, shared_yaxes="all", vertical_spacing = 0.015, subplot_titles=list(dictionary.keys()))

for index, item in enumerate(dictionary):
    dataframe = dictionary.get(item)
    if index == 0:
      traces = processing.trace_dataframe_list([dataframe], sensor_names, True)
    else:
      traces = processing.trace_dataframe_list([dataframe], sensor_names, False)
    for trace in traces:
      fig.add_trace(trace, row=index+1, col=1)

fig.update_layout(margin_l=0, margin_r=0, height=1000)  # no margins left or right, bigger height
fig.update_layout(yaxis_title="acceleration")

fig.show()
```

# 5. Comparison of pairs with window calculation

```python
import itertools

def get_comparison(a, b):
  output = pd.DataFrame()
  output["score_add"] = a["score_clean"] + b["score_clean"]
  mask = processing.calc_windows_between_two_dataframes(a["identifier"], b["identifier"])
  output["score"] = output["score_add"] * mask
  # output["mask"] = mask
  return output

jogging_comparison = dict()
for a, b in itertools.combinations(dictionary, 2):
  jogging_comparison[a, b] = get_comparison(dictionary[a], dictionary[b])

for key in jogging_comparison:
  df = jogging_comparison.get(key)
  df["score_bin"] = df["score"]
  df["score_bin"].loc[df["score_bin"] < 0.5] = 0
  df["score_bin"].loc[df["score_bin"] >= 0.5] = 1
  jogging_comparison[key] = df

keys = []
for key in jogging_comparison.keys():
  keys.append(''.join(key))
sensor_names = ["score", "score_add", "score_bin"]
fig = make_subplots(rows=len(jogging_comparison), cols=1, shared_xaxes=True, shared_yaxes="all", vertical_spacing = 0.005, subplot_titles=keys)

for index, item in enumerate(jogging_comparison):
    dataframe = jogging_comparison.get(item)
    if index == 0:
      traces = processing.trace_dataframe_list([dataframe], sensor_names, True)
    else:
      traces = processing.trace_dataframe_list([dataframe], sensor_names, False)
    for trace in traces:
      fig.add_trace(trace, row=index+1, col=1)

fig.update_layout(margin_l=0, margin_r=0, height=1600)  # no margins left or right, bigger height
fig.update_layout(yaxis_title="acceleration")

fig.show()
```

# 6. Analysis of detection

```python
import numpy as np

dictionary = jogging_comparison

def calc_detection_percent_of_dict_within_timeframe(dictionary, row, start=None, end=None):
  result = pd.DataFrame()
  result.loc[:, [row]] = np.nan

  for key in dictionary:
    df = dictionary[key].fillna(0)
    if start is None:
      df = df.loc[:end]
    if end is None:
      df = df.loc[start:]
    else:
      df = df.loc[start:end]

    sum_1 = (df[row] == 1).values.sum()
    sum_0 = (df[row] == 0).values.sum()
    result.loc[''.join(key)] = (sum_1/(sum_0+sum_1)) * 100

  return result

start1 = "2022-09-30 05:40:00"
end1 = "2022-09-30 06:04:00"
start2 = "2022-09-30 06:10:00"
end2 = "2022-09-30 06:32:00"

result = pd.DataFrame()
result["activity1"] = calc_detection_percent_of_dict_within_timeframe(dictionary, "score_bin", start=start1, end=end1)
result["activity2"] = calc_detection_percent_of_dict_within_timeframe(dictionary, "score_bin", start=start2, end=end2)
result["before"] = calc_detection_percent_of_dict_within_timeframe(dictionary, "score_bin", end=start1)
result["between"] = calc_detection_percent_of_dict_within_timeframe(dictionary, "score_bin", start=end1, end=start2)
result["after"] = calc_detection_percent_of_dict_within_timeframe(dictionary, "score_bin", start=end2)
result["combined"] = (result["before"] + result["between"] + result["after"]) / 3
result["activity_combined"] = (result["activity1"] + result["activity2"]) /2

print("Percentage in range: " + str(result["activity_combined"].mean()) + " %")
print("Percentage outside range: " + str(result["combined"].mean()) + " %")
print()

result_without_0 = result[result["activity_combined"] > 0]
print("Result without pairs where no activity was detected:")
print("Percentage in range: " + str(result_without_0["activity_combined"].mean()) + " %")
print("Percentage outside range: " + str(result_without_0["combined"].mean()) + " %")
```

# Calculate lag of every item with each other

```python
import itertools
from scipy import signal
import pandas as pd

jogging_lag = dict()
mean = 0
count = 0
for a, b in itertools.permutations(jogging_vec, 2):
    lag =  processing.calc_lag(jogging_vec[a], jogging_vec[b])
    jogging_lag[a, b] = lag
    mean = mean + abs(lag)
    count = count + 1
mean = mean/count
mean = mean * 0.02
print("Mean Lag: " + str(mean) + "s")
print()

lag_df = pd.DataFrame(index=jogging_vec.keys(), columns=jogging_vec.keys())
for keys in jogging_vec:
  for keys2 in jogging_vec:
    if (keys == keys2):
      lag_df[keys][keys2] = 0
    else:
      lag_df[keys][keys2] = jogging_lag[keys, keys2]

lag_df = lag_df * 0.02
print(lag_df)

# from google.colab import files
# lag_df.to_excel("lag_df.xlsx")
# files.download("lag_df.xlsx")
```
